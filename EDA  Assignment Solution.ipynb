#### Checking missing value percentage 

```python
def null_percentage(df):
    output = round(df.isnull().sum()/len(df.index)*100,2)
    return output
```

```python
NUll_col = null_percentage(BankapplData)
NUll_col
```

```python
plt.figure(figsize=[10,8])
NUll_col.plot.hist().set(xlabel='Percentage Missing Values',ylabel='Number of Columns',title='Distribution of Missing Values across Columns')
plt.show()
```

```python
# filter out the columns have null values greater than 40 percentage
NUll_col =NUll_col[NUll_col>40]
```

```python

plt.figure(figsize = (10,8), dpi=500)
NUll_col.plot(kind = "barh")
plt.title("Null values in columns")
plt.xlabel('Percentage of Null')
plt.show()
```

##### From the above graph it is evident that there are many columns with the missing values percentage more than 40%, since 40% is a significant number for missing values in the dataset, we can drop these columns.

```python
#Dropping Columns with Missing Values
#List of columns with missing values percentage greater than 40

droped_cols = NUll_col.index.tolist()

print('Number of Columns that have missing values percentage greater than 40%:', len(droped_cols))
```


#### `Note :`It can be seen most of the families has 2 family members, hence it is safe to impute missing values with Mode of CNT_FAM_MEMBERS column


```python
#EXT_SOURCE_3,EXT_SOURCE_2 are normalized scores from data source 3 and 2, we can impute missing values with average value
fill_null('EXT_SOURCE_3',BankapplData,BankapplData.EXT_SOURCE_3.mean)
fill_null('EXT_SOURCE_2',BankapplData,BankapplData.EXT_SOURCE_2.mean)
#OCCUPATION_TYPE column has unordered categorical values, we can impute missing values with a value 'Unavailable'
fill_null('OCCUPATION_TYPE',BankapplData,'Unavailable')
#CNT_FAM_MEMBERS can be imputed with the mode
fill_null('CNT_FAM_MEMBERS',BankapplData,BankapplData.CNT_FAM_MEMBERS.mode)
```
